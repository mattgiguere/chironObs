I'm going to make a pandas dataFrame that mirrors that format. First, I'll restore a bit of the objects table into a pandas dataframe to check it looks like I would expect it to look. 

```python
def connectChiron():
    """connect to the database"""
    #retrieve credentials:
    cmd = 'echo $AeroFSdir'
    #read in the AeroFSdir string and
    #chop off the newline character at the end
    cdir = subprocess.check_output(cmd, shell=True)
    cdir = cdir[0:len(cdir)-1]
    credsf = open(cdir+'.credentials/SQL/ceaye', 'r')
    creds = credsf.read().split('\n')
    conn = pymysql.connect(host=creds[0],
                           port=int(creds[1]),
                           user=creds[2],
                           passwd=creds[3],
                           db=creds[4])
    #cur = conn.cursor()
    return conn
```

```python
conn = connectChiron()
```
